```python
df['diagnosis']=df['diagnosis'].map({'M':1,'B':0})
df
```

```python
X =df.drop(['diagnosis'],axis =1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify = y,train_size =0.8,random_state = 12345)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

```python
X_train_n= X_train_sc.T
X_test_n = X_test_sc.T
y_train_n = y_train.values.reshape(1,-1)
y_test_n = y_test.values.reshape(1,-1)
# rows are the features
```

```python
print(f"train shape {X_train_n.shape} AND {y_train_n.shape}")
```

```python
print(f"test shape {X_test_n.shape} AND {y_test_n.shape}")
```

```python
import scipy as sc
```

```python
sc.special.expit(0.5) #sigmoid function 1/1+exp(-x)
```

```python
# Initialize weight to zero

w = np.zeros((X_train_n.shape[0],1))
```

```python
w.shape
```

```python
b = 0
alpha = 0.01 #learning rate
num_iters = 100 #also known as epochs
```

```python
costs = []
m = X_train_n.shape[1]
```

```python
m
```

```python
for i in range(num_iters):
    print(f"Iteration {i}")
     ## Calculate prediction based on the weights
    ## Sigmoid(w.T*X+b)
    A = sc.special.expit(np.dot(w.T,X_train_n)+b) ## Forward Propagati
      # Compute cost
    cost = np.sum(((- np.log(A))*y_train_n + (-np.log(1-A))*(1-y_train_n)))/m
       
    #backward propagation
    
    dw = (np.dot(X_train_n,(A-y_train_n).T))/m ## dJ/dw
    db = (np.sum(A-y_train_n))/m ## dJ/db
          
    # Gradient Descent
    
    w = w - alpha*dw 
    b = b - alpha*db
    
    costs.append(cost)
```

```python
w
```

```python
b
```

```python
nn_tr_predict = sc.special.expit(np.dot(w.T,X_train_n)+b) > 0.5 # Training set prediction
nn_ts_predict = sc.special.expit(np.dot(w.T,X_test_n)+b) > 0.5 # Test set predictio
```

```python
nn_ts_predict*1
```

```python
nn_tr_predict*1
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
# Accuracy on training and test set
acc_tr = accuracy_score(y_pred = nn_tr_predict[0], y_true = y_train_n[0])
acc_ts = accuracy_score(y_pred = nn_ts_predict[0], y_true = y_test_n[0])
```

```python
print(acc_tr,acc_ts)
```

```python
# Training set confusion matrix
confusion_matrix(y_pred = nn_tr_predict[0], y_true = y_train_n[0])

```

```python
# Plot how the cost keeps reducing
import matplotlib.pyplot as plt
plt.figure(figsize = (12,8))
plt.plot(costs)
plt.xlabel("Number of Iterations")
plt.ylabel("Costs")
plt.title("Costs vs Iterations")
plt
```
